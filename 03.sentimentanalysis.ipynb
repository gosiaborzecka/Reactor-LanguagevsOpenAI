**Azure AI Language call**

```python
from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential

# Set up your Azure Cognitive Services credentials
key = "{AzureLanguageServicesKey}"
endpoint = "{AzureLanguageServicesEndpoint}"

# Create a TextAnalyticsClient
credential = AzureKeyCredential(key)
client = TextAnalyticsClient(endpoint, credential)

# Perform sentiment analysis
for text in text_to_analyse:
    result = client.analyze_sentiment([text])[0]

    # Get the sentiment score
    sentiment_score = result.confidence_scores

    if (result.sentiment == "positive"):
        text_language.append(f"Positive {sentiment_score.positive * 100}%")
    elif (result.sentiment == "negative"):
        text_language.append(f"Negative {sentiment_score.negative * 100}%")
    elif (result.sentiment == "neutral"):
        text_language.append(f"Neutral {sentiment_score.neutral * 100}%")
    else: 
        text_language.append(f"Mixed {sentiment_score.positive * 100}%, {sentiment_score.negative * 100}%, {sentiment_score.neutral * 100}%")

    # # Print the sentiment score
    print(text)
    print("Sentiment score: {}".format(sentiment_score))
    print(result.sentiment)
    print("****")
```

**Open AI call with text results**

```python
from openai import OpenAI

client = OpenAI(
    # defaults to os.environ.get("OPENAI_API_KEY")
    api_key="{OPENAI_API_KEY}",
)


for text in text_to_analyse:

    completion = client.chat.completions.create(
        model="gpt-4",
        messages = [
            {"role": "system", "content": "You are an AI language model trained to analyze and detect the sentiment of product reviews."},
            {"role": "user", "content": f"Analyze the following product review and determine if the sentiment is: positive, negative or neutral and show precent of confidence. Return only a single word, either POSITIVE, NEGATIVE, NEUTRAL or MIXED with the precent of confidence: {text}"}
        ]
    )
    text_openAI.append(completion.choices[0].message.content)
    print(text)
    print(completion.choices[0].message.content)
    print("***")
```

**Open AI call with JSON results**

```python
from openai import OpenAI
import json

client = OpenAI(
    # defaults to os.environ.get("OPENAI_API_KEY")
    api_key="{OPENAI_API_KEY}",
)


for text in text_to_analyse:

    completion = client.chat.completions.create(
        model="gpt-3.5-turbo-1106",
        response_format={ "type": "json_object" },
        messages = [
            {"role": "system", "content": "You are an AI language model trained to analyze and detect the sentiment of product reviews."},
            {"role": "user", "content": f"Analyze the following product review and determine if the sentiment is: positive, negative or neutral and show precent of confidence. Return only a single word, either POSITIVE, NEGATIVE, NEUTRAL or MIXED with the precent of confidence and return in JSON format: {text}"}
        ]
    )



    python_object = json.loads(completion.choices[0].message.content)
    sentiment = python_object["sentiment"]
    confidence = python_object["confidence"]
    text_openAI_json.append(str(sentiment) + " " + str(confidence))
    
    print(text)
    print(completion.choices[0].message.content)
    print("***")
```
